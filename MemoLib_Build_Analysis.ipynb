## Architecture Overview
**Full-Stack Application**: Next.js 16 (TypeScript, Server Components) + Python Backend (FastAPI/Flask)

- **Frontend**: `/src/frontend` - Next.js 16, App Router, 1600+ npm packages
- **Backend**: `/src/backend` (FastAPI) + `/backend-python` (Flask dev, port 5000)
- **Database**: Prisma (Next.js) + SQLAlchemy (Python)
- **Monitoring**: Sentry + OpenTelemetry instrumentation
- **Deployment**: Vercel, Azure SWA, Docker, Railway, Render

## ğŸ—ï¸ Build Pipeline Phases

### Phase 1: Install Dependencies
```bash
npm install --legacy-peer-deps  # Frontend (with Prisma generate)
pip install -r requirements-python.txt  # Backend
```

### Phase 2: Lint & Type Check
```bash
npm run lint                    # ESLint (50 warnings max)
npm run type-check             # TypeScript strict
flake8 .                        # Python linting
```

### Phase 3: Run Tests
```bash
npm test                        # Jest (parallel, 50% workers)
pytest -v --tb=short           # Python tests
npm run test:e2e               # Playwright E2E
```

### Phase 4: Production Build
```bash
NODE_OPTIONS=--max-old-space-size=8192 next build --turbo
# Output mode: export (Azure), standalone (Docker), undefined (Vercel/Windows)
```
